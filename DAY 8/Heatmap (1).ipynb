```python
#Heatmap:-- it is based on correlation between datapoints and which lies between -1 to 1
import seaborn as sns
s1=sns.load_dataset("flights")
print(s1.head(5))
print(s1.tail(5))
```

```python
flights_corr=s1[["year","passengers"]]
flights_corr
```

```python
corr=flights_corr.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
# heatmap using pivot table:--
s2=s1.pivot_table(values="passengers",index="month",columns="year")
s2
```

```python
sns.heatmap(s2,annot=False,cbar=True,cmap='viridis',linewidth=0.5,linecolor="y")
```
